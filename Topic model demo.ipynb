```python
textdir = "/Users/erinlane/Desktop/Political Tweets/" #type the file directory where your corpus is located
dirliststart = sorted(os.listdir(textdir))
dirlist = []
for filename in dirliststart:
    if filename[-4:] == ".txt":
        dirlist.append(filename)

lemmatized_texts = []
for file in dirlist:
    with open(textdir + file, "r") as f:
        lemmatized_texts.append(f.read())
```

```python
def gen_words(texts):
    final = []
    for text in texts:
        new = gensim.utils.simple_preprocess(text, deacc=True)
        final.append(new)
    return (final)

data_words = gen_words(lemmatized_texts) 

print (data_words[0][0:50]) 
```

```python
def gen_words(texts):
    final = []
    for text in texts:
        new = gensim.utils.simple_preprocess(text, deacc=True)
        final.append(new)
    return (final)

data_words = gen_words(lemmatized_texts)

print (data_words[0][0:20])
```

```python
id2word = corpora.Dictionary(data_words)

corpus = []
for text in data_words:
    new = id2word.doc2bow(text)
    corpus.append(new)

print (corpus[0][0:20])

word = id2word[[0][:1][0]]
print (word)
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=20,
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha="auto")
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word, mds="mmds", R=30)
vis
```
